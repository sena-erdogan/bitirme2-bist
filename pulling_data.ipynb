```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Define the search parameters
query = 'coronavirus'
language = 'tr'
since_date = '2010-01-01'
until_date = '2024-12-31'

# Use snscrape to scrape the data
url = f'https://eksisozluk.com/{query}--{language}?a=daterange:{since_date},{until_date}'
tweets = list(sntwitter.TwitterSearchScraper(url).get_items())

# Process the scraped data
results = []
for tweet in tweets:
    text = tweet.content
    date = tweet.date.strftime('%Y-%m-%d %H:%M:%S')
    results.append({'text': text, 'date': date})

# Create a pandas dataframe to store the results
df = pd.DataFrame(results)

# Print the dataframe
print(df)
```
